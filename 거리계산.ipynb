```python
import pandas as pd
import numpy as np
from haversine import haversine
import time
from tqdm import tqdm

distance = pd.read_csv("C:\\Users\\USER\\Desktop\\제주전기차_수정\\전처리데이터.csv", engine='python')
```

```python
distance.fillna(0, inplace=True)
```

```python
distance.head(10)
```

```python
distance['거리_전기차'] = 0
distance['거리_충전소'] = 0
```

```python
distance2=distance[distance['전기차']!=0]
distance2.head(5)
```

```python
for j in tqdm(range(0, len(distance))):
    for i in (range(0, len(distance2))):
        lon = distance['lon'][j]
        lat = distance['lat'][j]
        lon2 = distance2['lon'][i].reset_index(drop=True)
        lat2 = distance2['lat'][i].reset_index(drop=True)
        a = (lat, lon)
        b = (lat2, lon2)
        c = haversine(a, b) * 1000
        if c <= 500:
             distance['거리_전기차'][j] = 1
```

```python
distance3=distance[distance['충전소']!=0].reset_index(drop=True)
distance3.head(5)
```

```python
for j in tqdm(range(0, len(distance))):
    for i in range(0, len(distance3)):
        lon = distance['lon'][j]
        lat = distance['lat'][j]
        lon2 = distance3['lon'][i]
        lat2 = distance3['lat'][i]
        a = (lat, lon)
        b = (lat2, lon2)
        c = haversine(a, b) * 1000
        if c <= 500:
             distance['거리_충전소'][j] = 1
```

```python
distance4=distance[distance['교통량']!=0].reset_index(drop=True)
distance4.head(5)
```

```python
for j in tqdm(range(0, len(distance))):
    for i in range(0, len(distance4)):
        lon = distance['lon'][j]
        lat = distance['lat'][j]
        lon2 = distance4['lon'][i]
        lat2 = distance4['lat'][i]
        a = (lat, lon)
        b = (lat2, lon2)
        c = haversine(a, b) * 1000
        if c <= 500:
             distance['거리_교통량'][j] = 1
```
